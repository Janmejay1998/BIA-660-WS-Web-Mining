```python
import tomotopy as tp
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize,word_tokenize
import re
import csv
```

```python
def tokenizer(doc,sw):
    return [word for word in [re.sub('[^a-z]','',x.lower()) for x in doc.strip().split()] if word not in sw and len(word)>2]
```

```python
# Function to return list of topics with top 15 features
def func(num):
    input_path ='hotelreviews.csv'
    fin=open(input_path,encoding='utf8')
    # Reads the data
    #reader=csv.reader(fin)
    reader = list(csv.reader(open('hotelreviews.csv'), delimiter='\t'))
    docx = []
    
    mdl = tp.LDAModel(k=num)
    sw=stopwords.words('english')
    for line in reader:
            # Get the text and rating
            text,rating=line
            #print("text : ",text, "\n","rating : ",rating,"\n")
            docx.append(text)

    for doc in docx:
            mdl.add_doc(tokenizer(doc,sw))

    for i in range(0, 500,10):
        mdl.train(10)
        print('Iteration: {}\tLog-likelihood: {}'.format(i, mdl.ll_per_word))
    
    topix = []
    for k in range(mdl.k):
        topk_words=[pair[0] for pair in mdl.get_topic_words(k, top_n=15)]
        print(k,topk_words,"\n")
        topix.append(topk_words)
    # List Combiner test
    lis = []
    for i in topix:
        topic_str = ''
        for j in i:
            topic_str +=  j + ','
        topic_str = topic_str[:-1]

        lis.append(topic_str)
        
    return lis
       
```

```python
# Function to write to csv
def writer_csv(lis):
    # creating csv and adding elements 
    with open('topics3.csv', 'w', encoding='UTF8',newline='') as f:
        # create the csv writer
        writer = csv.writer(f,delimiter = ',')

        # write a row to the csv file
        for i in range(len(lis)):
            writer.writerow([lis[i]])
    print("Completed !")
```
