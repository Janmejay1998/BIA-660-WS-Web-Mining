```python
import tomotopy as tp
from nltk.corpus import stopwords
import csv
import re
```

#### Function for tokenizing

```python
def tokenizer(doc,sw):
    return [word for word in [re.sub('[^a-z]','',x.lower()) for x in doc.strip().split()] if word not in sw and len(word)>2]
```

#### Function for returning a list of topics with top 15 features

```python
def topic_func(num):

    # Stopwords
    sw = stopwords.words('english')

    # New LDA model
    mdl = tp.LDAModel(k=num)

    with open('hotelreviews.csv', newline='') as f:
        reader = csv.reader(f)
        for row in reader:
            doc = row[0]
            doc = doc.rsplit('\t', 1)[0]
            mdl.add_doc(tokenizer(doc,sw))        

    # Train LDA model
    for i in range(0, 500,10):
        mdl.train(10)
        print('Iteration: {}\tLog-likelihood: {}'.format(i, mdl.ll_per_word))  

    # Print doc info
    # for doc in mdl.docs:
    #    print ([round(p,2) for p in doc.get_topic_dist()])
    #    print()          
    
    topics_list = []
    # Print topic info
    for k in range(mdl.k):
        topk_words=[pair[0] for pair in mdl.get_topic_words(k, top_n=15)]
        print(k,topk_words,"\n")
        topics_list.append(topk_words)

    # Storing sets of topics into list
    ls = []
    for i in topics_list:
        topic_list_to_str = ''
        for j in i:
            topic_list_to_str +=  j + ','
        topic_list_to_str = topic_list_to_str[:-1]
        ls.append(topic_list_to_str)
        
    return ls
```

#### Function for writing list of topics into csv file

```python
def writer_csv(topic_list): 
    with open('topics.csv', 'w', encoding="utf-8", newline='') as fw:   # Creating csv file and adding elements to it
        writer = csv.writer(fw, delimiter = ',')    # Creating the csv writer
        for i in range(len(topic_list)):    # Writing a row to the csv file
            writer.writerow([topic_list[i]])
    print("LDA Topic Model Successfully Finished!")
```
